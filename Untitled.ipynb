```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
file_path = 'Electric_Vehicle_Population_Size_History_By_County_.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Get a concise summary of the dataframe
data.info()
```

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Convert 'Electric Vehicle (EV) Total' to numeric, coerce errors to NaN
data['Electric Vehicle (EV) Total'] = pd.to_numeric(data['Electric Vehicle (EV) Total'], errors='coerce')

# Check for missing values
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])

# Drop rows with missing values, if necessary
data.dropna(inplace=True)

# Recheck the info to confirm changes
data.info()
```

```python
# Summary statistics for Electric Vehicle (EV) Total
ev_summary_stats = data['Electric Vehicle (EV) Total'].describe()
print(ev_summary_stats)
```

```python
import matplotlib.pyplot as plt

# Group data by Date and sum up the Electric Vehicle totals
trend_data = data.groupby('Date')['Electric Vehicle (EV) Total'].sum().reset_index()

# Plotting the trend of Electric Vehicle totals over time
plt.figure(figsize=(12, 6))
plt.plot(trend_data['Date'], trend_data['Electric Vehicle (EV) Total'], marker='o', linestyle='-')
plt.title('Trend of Electric Vehicle (EV) Total Over Time')
plt.xlabel('Date')
plt.ylabel('Sum of Electric Vehicle (EV) Total')
plt.grid(True)
plt.show()
```

```python
# Group data by County and sum up the Electric Vehicle totals, then get the top 10 counties
county_ev_totals = data.groupby('County')['Electric Vehicle (EV) Total'].sum().sort_values(ascending=False).head(10)

# Plotting the top counties by Electric Vehicle totals
plt.figure(figsize=(12, 6))
county_ev_totals.plot(kind='bar', color='cadetblue')
plt.title('Top 10 Counties by Electric Vehicle (EV) Total')
plt.xlabel('County')
plt.ylabel('Sum of Electric Vehicle (EV) Total')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group data by County and calculate the mean percentage of electric vehicles
county_ev_percentage = data.groupby('County')['Percent Electric Vehicles'].mean().sort_values(ascending=False).head(10)

# Plotting the top counties by average percentage of Electric Vehicles
plt.figure(figsize=(12, 6))
county_ev_percentage.plot(kind='bar', color='coral')
plt.title('Top 10 Counties by Average Percentage of Electric Vehicles')
plt.xlabel('County')
plt.ylabel('Average Percentage of Electric Vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Combine the total counts and percentage data into one DataFrame
summary_table = pd.DataFrame({
    'Sum of EV Total': county_ev_totals,
    'Average Percentage of EVs': county_ev_percentage
}).reset_index()

# Display the summary table
print(summary_table)
```

```python
# Filtering counties that are present in both total counts and percentage datasets
common_counties = county_ev_totals.index.intersection(county_ev_percentage.index)

# Creating the summary table with common counties
summary_table = pd.DataFrame({
    'Sum of EV Total': county_ev_totals[common_counties],
    'Average Percentage of EVs': county_ev_percentage[common_counties]
}).reset_index()

# Display the summary table
print(summary_table)

```

```python
# Join the two Series into a single DataFrame on the index which represents the counties
summary_table = pd.merge(
    county_ev_totals.reset_index(),
    county_ev_percentage.reset_index(),
    on='County',
    how='outer'
)

# Rename the columns appropriately
summary_table.columns = ['County', 'Sum of EV Total', 'Average Percentage of EVs']

# Fill NaN values in 'Average Percentage of EVs' with the mean value
summary_table['Average Percentage of EVs'] = summary_table['Average Percentage of EVs'].fillna(
    summary_table['Average Percentage of EVs'].mean()
)

# Display the summary table
print(summary_table)
```

```python
# Save the summary table to a CSV file
summary_table.to_csv('summary_table.csv', index=False)
```

```python
# Define a function to find outliers using IQR
def find_outliers_IQR(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[((df[column] < (Q1 - 1.5 * IQR)) |(df[column] > (Q3 + 1.5 * IQR)))]
    return outliers

# Assuming you want to find outliers in the 'Percent Electric Vehicles' column
outliers_percent_ev = find_outliers_IQR(data, 'Percent Electric Vehicles')
print("Outliers in 'Percent Electric Vehicles':\n", outliers_percent_ev)
```
